```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix , precision_score,classification_report,roc_auc_score, roc_curve
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('..\data\exoplanents_cleaned_new.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
feature_cols = [
    'koi_score', 'koi_fpflag_nt', 'koi_fpflag_ss', 'koi_fpflag_co', 'koi_fpflag_ec',
    'koi_period', 'koi_time0bk', 'koi_impact', 'koi_duration', 'koi_depth',
    'koi_prad', 'koi_teq', 'koi_insol', 'koi_model_snr', 'koi_steff',
    'koi_slogg', 'koi_srad'
]
```

```python
X = df[feature_cols]
```

```python
y = df['koi_disposition'].apply(lambda x: 1 if x == 'CONFIRMED' else 0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, 
)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_prob = model.predict_proba(X_test)[:,1]
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
roc_auc_score(y_test, y_prob)
```
